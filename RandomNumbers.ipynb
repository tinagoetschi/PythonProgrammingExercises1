## Random numbers in Python

python has a LIBRARY (import random) for generating random (actually only pseudo random) numbers

```python
import random

secretNum= random.randint(1,10)

print(secretNum)

```

```python
from random import randint
num = randint(0, 100)

print('your random number is ', num)
```

```python
#get a random number in teh five times table
x = random.randrange(0,101,5)
print(x)
```
